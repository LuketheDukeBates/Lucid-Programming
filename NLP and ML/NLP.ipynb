```python
alice_words = nltk.corpus.gutenberg.words('carroll-alice.txt')
alice_words
```

```python
alice_chars = nltk.corpus.gutenberg.raw('carroll-alice.txt')
```

```python
alice_sents = nltk.corpus.gutenberg.sents('carroll-alice.txt')
alice_sents
```

```python
#Average word length
def averageWordLength(chars, words):
    return len(chars)//len(words)
averageWordLength(alice_chars, alice_words)
```

```python
#Average sentence length
def averageSentLength(words, sents):
    return len(words)//len(sents)
averageSentLength(alice_words, alice_sents)
```

```python
for file in nltk.corpus.gutenberg.fileids():
    chars = nltk.corpus.gutenberg.raw(file)
    words = nltk.corpus.gutenberg.words(file)
    sents = nltk.corpus.gutenberg.sents(file)
    
    print(file + " has an average word length of " + str(averageWordLength(chars, words)) + 
                                                         " and and average sentence length of "+
                                                        str(averageSentLength(words, sents)))
```

```python
for file in nltk.corpus.inaugural.fileids():
    print(file)
```

```python
for file in nltk.corpus.inaugural.fileids():
    america_count = 0
    for word in nltk.corpus.inaugural.words(file):
        if word.lower().startswith('america'):
            america_count += 1
    president = file[5:-4]
    year = file[:4]
    print("President " + president + " of year " + year + " said America " + str(america_count) + " times.")
```

```python
cfd = nltk.ConditionalFreqDist(
            (target, file[:4])
            for file in nltk.corpus.inaugural.fileids() 
            for word in nltk.corpus.inaugural.words(file)
            for target in ['america', 'citizen']
            if word.lower().startswith(target))
```

```python
plt.figure(figsize=(12,8))
cfd.plot()
```

### Stemming and lemmatization

```python
from nltk.stem import PorterStemmer
porter = PorterStemmer()
word_lst = ["connected", 'connecting', "connection", "connections"]
```

```python
for word in word_lst:
    print(porter.stem(word))
```

```python
word_lst = ["argue", "arguing", "argues", "arguing", "argued", "argus"]
```

```python
for word in word_lst:
    print(porter.stem(word))
```

```python
from nltk.stem import LancasterStemmer
from nltk.stem import SnowballStemmer
```

#### stemmers operate only on a singler word WITHOUT knowledge of the context and therefore cannot discriminate between words which have different meansings depending on the part of speech

```python
lancaster = LancasterStemmer()
snowball = SnowballStemmer(language = 'english')
```

```python
for word in word_lst:
    print(lancaster.stem(word))
```

```python
for word in word_lst:
    print(snowball.stem(word))
```
